## Primeiro acesso:
Casos de COVID19 nos munícipios do estado de São Paulo

Fonte: SEADE 
Formato: CSV

```
import requests
import io
import csv

url = "https://raw.githubusercontent.com/seade-R/dados-covid-sp/master/data/dados_covid_sp.csv"

data = requests.get(url).content

dados_covid_sp = csv.reader(io.StringIO(data.decode("utf-8")), delimiter= ';')

line_count = 0

for row in dados_covid_sp:

  if line_count == 0:
      print("Dados sobre COVID19 no estado de São Paulo:")

  elif row[0] != "Ignorado":
      print(f'{row[0]} possui {row[5]} casos de COVID-19.')

  line_count += 1
```

## Segundo acesso:

Munícipios no Estado de São Paulo

Fonte: IBGE \
Formato: JSON

```
import requests
import io
import json

url = "https://servicodados.ibge.gov.br/api/v1/localidades/estados/35/municipios"

data = requests.get(url).content

municipios_sp = json.loads(data)

print(json.dumps(municipios_sp, indent=4, sort_keys=True, ensure_ascii=False))
```
